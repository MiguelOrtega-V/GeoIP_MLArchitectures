## FASE 1 - Carga de datos

Los datos ya están divididos previamente en entrenamiento/producción

```python
import pandas as pd

dataset_path = '../../../Datasets/TrainingDatasets/new_measurements_learning.csv'

training_data = pd.read_csv(dataset_path)

print(f'El dataset tiene {training_data.shape[0]} filas y {training_data.shape[1]} columnas')

training_data.head()

```

```python
# Nos quedamos de training data por las entradas con un measure_id que aparezca 6 veces (los 6 monitores)
print(f"Número de entradas antes de eliminar medidas faltantes: {len(training_data)}")
training_data = training_data[training_data['measure_id'].map(training_data['measure_id'].value_counts()) == 6]
print(f"Número de entradas después de eliminar medidas faltantes: {len(training_data)}")
```

```python
# Representamos la cabecera
training_data.head()
```

```python
# Para estas primeras pruebas, nos vamos a quedar únicamente con los landmark_type ripe_anchor y ripe_probe
print(f"Número de entradas antes de eliminar dns y router intermedios: {len(training_data)}")
training_data = training_data[training_data['landmark_type'].isin(['ripe_anchor', 'ripe_probe'])]
print(f"Número de entradas después de eliminar dns y router intermedios: {len(training_data)}")
```

```python
# Guardamos el dataset en un archivo .csv en una carpeta temporal
training_data.to_csv('../tmp/training_data.csv', index=False)
```
